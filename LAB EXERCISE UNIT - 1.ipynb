```python
# LAB PROBLEM 1 
# Write a program take text file as an input and print word, character count and ascii value of each characters  
# as output. (Hint: Use open(), read() and split())
f =open("q1.txt","rt")
txt = f.read()
print(txt)
arr = []
arr2 = []
for i in txt:
    arr.append(i)
arr2 = [ord(j) for j in arr]
for i in range(len(arr2)):
    arr2[i] = arr2[i] + 4
    if arr2[i]>122:
        arr2[i] = ((arr2[i]%122)+96)
arr5 = [chr(i) for i in arr2]
entxt = ''.join(map(str, arr5))
print("The Encrypted text is:",entxt)
```

```python
# LAB PROBELM 3 
# Encryption of Plain Text and Decryption of cipher text using Vigenere Cipher
# Key: deceptivedeceptivedeceptive
# Plaintext: wearediscoveredsaveyourself
# Cipher text:ZICVTWQNGRZGVTWAVZHCQYGLMGJ
def encrypt(string, key):
    ciptext = []
    for i in range(len(string)):
        x = (ord(string[i]) + ord(key[i])) % 26
        x += ord('a')
        ciptext.append(chr(x))
    return "".join(ciptext)


def decrypt(entxt, key):
    plaintext = []
    for i in range(len(entxt)):
        y = (ord(entxt[i]) - ord(key[i]) + 26) % 26
        y += ord("A")
        plaintext.append(chr(y))
    return "".join(plaintext)


string = "wearediscoveredsaveyourself"
key = "deceptivedeceptivedeceptive"
entext = encrypt(string, key)
print("Encrypted Text " + entext)
encypttxt = "ZICVTWQNGRZGVTWAVZHCQYGLMGJ"
dectxt = decrypt(encypttxt, key)
print("Decrypted Text " + dectxt)
```

```python
# LAB PROBLEM 7 and 8 
# By using key matrix Raju want to send message “wearediscoveredsaveyourself” to Rani. 
# Can you build encryption process and find out what is the cipher text message send to Rani by using palyfair cipher.
# By using key matrix Raju sent encrypted message “LIIUDLTQNSLIZETQVTPKZEZFVBVZ” to Rani. 
# Can you build decryption process and find out what is the message send to Rani by using playfair cipher.
def generatingkey(key):
    L = list(dict.fromkeys(i for i in key))
    aplha = "ABCDEFGHIKLMNOPQRSTUVWXYZ"
    for i in aplha:
        if i not in L:
            L.append(i)
    key_matrix = [L[0:5], L[5:10], L[10:15], L[15:20], L[20:25]]
    return key_matrix


def message_to_diagraph(ptext):
    for s in range(0, len(ptext) + 1, 2):
        if s < len(ptext) - 1:
            if ptext[s] == ptext[s + 1]:
                ptext = ptext[:s + 1] + 'X' + ptext[s + 1:]
    if len(ptext) % 2 != 0:
        ptext = ptext[:] + 'X'
    return ptext


def getIndex(char, ctext):
    indexOfChar = []

    if char == "J":
        char = "I"

    for i, j in enumerate(ctext):
        for k, l in enumerate(j):
            if char == l:
                indexOfChar.append(i)
                indexOfChar.append(k)
                return indexOfChar


def encryption(ptext, key):
    cipherText = []
    keyMatrix = generatingkey(key)
    i = 0
    while i < len(ptext):
        n1 = getIndex(ptext[i], keyMatrix)
        n2 = getIndex(ptext[i + 1], keyMatrix)

        if n1[1] == n2[1]:
            i1 = (n1[0] + 1) % 5
            j1 = n1[1]

            i2 = (n2[0] + 1) % 5
            j2 = n2[1]
            cipherText.append(keyMatrix[i1][j1])
            cipherText.append(keyMatrix[i2][j2])

        elif n1[0] == n2[0]:
            i1 = n1[0]
            j1 = (n1[1] + 1) % 5
            i2 = n2[0]
            j2 = (n2[1] + 1) % 5
            cipherText.append(keyMatrix[i1][j1])
            cipherText.append(keyMatrix[i2][j2])

        else:
            i1 = n1[0]
            j1 = n1[1]

            i2 = n2[0]
            j2 = n2[1]

            cipherText.append(keyMatrix[i1][j2])
            cipherText.append(keyMatrix[i2][j1])
        i += 2
    return cipherText


def decryption(ctext, key):
    Text = []
    keyMatrix = generatingkey(key)
    i = 0
    while i < len(ctext):
        n1 = getIndex(ctext[i], keyMatrix)
        n2 = getIndex(ctext[i + 1], keyMatrix)

        if n1[1] == n2[1]:
            i1 = (n1[0] - 1) % 5
            j1 = n1[1]

            i2 = (n2[0] - 1) % 5
            j2 = n2[1]
            Text.append(keyMatrix[i1][j1])
            Text.append(keyMatrix[i2][j2])

        elif n1[0] == n2[0]:
            i1 = n1[0]
            j1 = (n1[1] - 1) % 5
            i2 = n2[0]
            j2 = (n2[1] - 1) % 5
            Text.append(keyMatrix[i1][j1])
            Text.append(keyMatrix[i2][j2])
            
        else:
            i1 = n1[0]
            j1 = n1[1]

            i2 = n2[0]
            j2 = n2[1]

            Text.append(keyMatrix[i1][j2])
            Text.append(keyMatrix[i2][j1])
        i += 2
    return Text


if __name__ == '__main__':
    key = "SRMAPUNIVERSITY"
    ptext = input("Enter plain text: ").replace(" ", "").upper()
    ctext = 'LIIUDLTQNSLIZETQVTPKZEZFVBVZ'
    print("Ciphered Text: " + "".join(encryption(message_to_diagraph(ptext), key)))
    print("Deciphered text from Rani: " + "".join(decryption(message_to_diagraph(ctext), key)))
```
